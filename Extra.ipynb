```python
from sklearn.model_selection import train_test_split  # type: ignore
import pandas as pd  # type: ignore
import numpy as np  # type: ignore

# Load the data
all_data = pd.read_csv("DATA/train.csv")

# Extract input columns and scale them by dividing by 255
X, Y = all_data.iloc[:, 1:].values / 255.0, all_data.iloc[:, 0].values

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3)
n = X_train.shape[0]

# Standard one-hot encoding
def one_hot(Y, num_classes):
    one_hot_Y = np.zeros((num_classes, Y.size))  # 10 rows and Y.size columns, each column represents all outputs for an image
                                                    # Y.size represents the batch size
    one_hot_Y[Y, np.arange(Y.size)] = 1  # Set the position corresponding to Y's value to 1
    return one_hot_Y

# Alternative one-hot encoding
def one_hot2(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    one_hot_Y = one_hot_Y.T  # Transpose to get the correct format
    return one_hot_Y

# Tests and printing
def test_one_hot():
    print("Original labels (Y):", Y_train[:5])

    # Apply one_hot
    one_hot_Y = one_hot(Y_train[:5], 10)  # Assuming 10 classes (0 to 9)
    print("\nOne-hot (using one_hot):")
    print(one_hot_Y)

    # Apply one_hot2
    one_hot_Y2 = one_hot2(Y_train[:5])
    print("\nOne-hot (using one_hot2):")
    print(one_hot_Y2)

# Run tests for one_hot and one_hot2
test_one_hot()
```

```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cargar el dataset MNIST
mnist = fetch_openml('mnist_784', version=1)
X, Y = mnist['data'], mnist['target'].astype(int)

# Seleccionar una muestra de 3000 imágenes para evitar sobrecarga en la correlación
X_sample = X[:3000].T  # Transponer para correlacionar columnas

# Calcular la matriz de correlación sobre la muestra seleccionada
correlation_matrix = np.corrcoef(X_sample)

# Crear el heatmap con un tamaño adecuado para que se vea más claro
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', cbar=True, square=True,
            xticklabels=10, yticklabels=10, vmin=-1, vmax=1, linewidths=0)
plt.title("Correlación entre columnas de una muestra de píxeles del MNIST")
plt.show()
```

```python
from sklearn.datasets import fetch_openml
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Establecer el tema de Seaborn
sns.set_theme(style="whitegrid")

# Cargar el dataset MNIST y seleccionar una submuestra de 100 píxeles
mnist = fetch_openml('mnist_784', version=1)
X = mnist['data'] / 255.0
X_sample = X.iloc[:, :80]  # Seleccionamos solo las primeras 100 columnas de píxeles

# Calcular la matriz de correlación y convertirla a formato largo
corr_mat = X_sample.corr().stack().reset_index(name="correlation")
corr_mat.columns = ["level_0", "level_1", "correlation"]

# Dibujar cada celda como un punto con tamaño y color variables
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8)
)

# Ajustes finales en el gráfico
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)

plt.show()
```
